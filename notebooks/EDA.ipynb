```python
#Juntando os dataframes para fazer a tabela fato

# Colunas escolhidas: partida_id | rodata | ano | data | arena | clube | condicao | adversario | vencedor | resultado | gols_marcados | gols_sofridos | saldo_gols | aproveitamento |
# formacao_mandante | formacao_visitante | chutes | chutes_no_alvo | precisao_chutes | posse_de_bola | passes | precisao_passes | faltas | impedimentos | escanteios | total_amarelos |
# total_vermelhos | mandante_Estado | visitante_Estado
```

```python
#Realizando o merge dos dataframes para criar a tabela fato
df_campfull_limpo = df_campfull.drop(columns=['rodata'])

df_fato = df_campestatistica \
    .merge(df_campfull_limpo, left_on='partida_id', right_on='ID', how='left') \
    .merge(df_cartoes_partida, on=['partida_id', 'clube'], how='left')
df_fato.columns
```

```python
#Mantendo apenas as colunas escolhidas para a análise
colunas_para_manter = ['partida_id', 'rodata', 'data','mandante', 'arena', 'clube', 'vencedor', 'visitante',
                        'chutes', 'chutes_no_alvo', 'posse_de_bola', 'passes', 'precisao_passes', 'faltas', 'impedimentos', 'escanteios', 'Amarelo',
                        'Vermelho', 'mandante_Placar', 'visitante_Placar']

df_fato = df_fato[colunas_para_manter]
df_fato.head()
```

```python
#categorizando as colunas

df_fato.info()
```

```python
df_fato['data'] = df_fato['data'].astype('datetime64[ns]')
df_fato['Amarelo'] = df_fato['Amarelo'].astype('Int64')
df_fato['Vermelho'] = df_fato['Vermelho'].astype('Int64')
```

```python
#Feature engineering: Criando colunas de  ano, saldo de gols, resultado, condiçao, aproveitamento, precisaod de chutes, flag_expulsao

df_fato['ano'] = pd.to_datetime(df_fato['data']).dt.year

# Gols do ponto de vista do clube
df_fato['gols_marcados'] = df_fato.apply(
    lambda x: x['mandante_Placar'] if x['clube'] == x['mandante'] else x['visitante_Placar'], axis=1
)
df_fato['gols_sofridos'] = df_fato.apply(
    lambda x: x['visitante_Placar'] if x['clube'] == x['mandante'] else x['mandante_Placar'], axis=1
)

df_fato['saldo_de_gols'] = df_fato['gols_marcados'] - df_fato['gols_sofridos']

df_fato['resultado'] = df_fato.apply(
    lambda x: 'Vitoria' if x['clube'] == x['vencedor']
    else ('Empate' if x['vencedor'] == '-' else 'Derrota'), axis=1
)

df_fato['condicao'] = df_fato.apply(
    lambda x: 'Mandante' if x['clube'] == x['mandante'] else 'Visitante', axis=1
)

df_fato['aproveitamento'] = df_fato['resultado'].map(
    {'Vitoria': 3, 'Empate': 1, 'Derrota': 0}
)

df_fato['precisao_chutes'] = (df_fato['chutes_no_alvo'] / df_fato['chutes']).round(2)

df_fato['flag_expulsao'] = df_fato['Vermelho'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# Renomeando e organizando o DF
df_fato = df_fato.rename(columns={
    'partida_id'      : 'id_partida',
    'rodata'          : 'rodada',
    'mandante_Placar' : 'placar_mandante',
    'visitante_Placar': 'placar_visitante',
    'Amarelo'         : 'total_amarelos',
    'Vermelho'        : 'total_vermelhos',
    'flag_expulsao'   : 'teve_expulsao',
})

df_fato = df_fato[[
    'id_partida', 'rodada', 'ano', 'data', 'arena',

    'clube', 'condicao', 'mandante', 'visitante', 'vencedor',

    'resultado', 'aproveitamento',
    'gols_marcados', 'gols_sofridos', 'saldo_de_gols',
    'placar_mandante', 'placar_visitante',

    'chutes', 'chutes_no_alvo', 'precisao_chutes',
    'posse_de_bola', 'passes', 'precisao_passes',
    'faltas', 'impedimentos', 'escanteios',

    'total_amarelos', 'total_vermelhos', 'teve_expulsao',
]]

df_fato.head()
```

```python
#Ajustando erros de dados inconsistentes
df_fato['precisao_chutes'] = df_fato['precisao_chutes'].fillna(0)

df_fato['total_amarelos'] = df_fato['total_amarelos'].fillna(0)

df_fato['total_vermelhos'] = df_fato['total_vermelhos'].fillna(0)


df_fato['precisao_passes'] = df_fato.apply(
    lambda x: df_fato[df_fato['clube'] == x['clube']]['precisao_passes'].mean().round(2)
    if x['passes'] == 0 else x['precisao_passes'], axis=1
)
```
