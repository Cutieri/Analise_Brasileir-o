```python
import pandas as pd
```

```python
path_cartoes = r'E:\Analise Brasileirao\Data\Raw\campcartoes.csv'
path_estatisticas = r'E:\Analise Brasileirao\Data\Raw\campestatisticas.csv'
path_campfull = r'E:\Analise Brasileirao\Data\Raw\campfull.csv'
```

```python
df_cartoes = pd.read_csv(path_cartoes, encoding="UTF-8", sep=",")
df_estatisticas = pd.read_csv(path_estatisticas, encoding="UTF-8", sep=",")
df_campfull = pd.read_csv(path_campfull, encoding="UTF-8", sep=",")
```

```python
df_cartoes.head()
```

```python
df_cartoes.info()
```

```python
df_cartoes['num_camisa'] = df_cartoes['num_camisa'].astype('Int64')
```

```python
#Setando tabela somente de 2020 para frente
df_cartoes = df_cartoes[df_cartoes['partida_id'] >= 6888]

df_cartoes.head()
```

```python
#Verificando se existem valores nulos
df_cartoes.isnull().sum()  
```

```python
#Exibindo as linhas onde o nome do atleta é nulo
df_cartoes[df_cartoes['atleta'].isnull()]
```

```python
#Removendo as linhas onde o nome do atleta é nulo - pois sao apenas 4 linhas, e nao tem como recuperar essa informacao
df_cartoes = df_cartoes.dropna(subset=['atleta'])
df_cartoes.isnull().sum()
```

```python
df_estatisticas.head()

#Setando tabela somente de 2020 para frente
df_estatisticas = df_estatisticas[df_estatisticas['partida_id'] >= 6888]
```

```python
df_estatisticas.head()
```

```python
df_estatisticas.info()
```

```python
#Recategorizando colunas posse de bola e precisao de faltas
# Remove o % e converte pra float
df_estatisticas['posse_de_bola'] = df_estatisticas['posse_de_bola'].str.replace('%', '').astype('Int64')
df_estatisticas['precisao_passes'] = df_estatisticas['precisao_passes'].str.replace('%', '').astype('Int64')
```

```python
#Verificando se existem valores nulos
df_estatisticas.isnull().sum()
```

```python
#Verificando as linhas onde há valor nulo
df_estatisticas[df_estatisticas['posse_de_bola'].isnull()]
```

```python
#Como são muitoas linhas, e não tem como recuperar essa informação, vamos preencher com a média da posse de bola de cada time

#Calculando a média da posse de bola por time
media_posse_de_bola = df_estatisticas.groupby('clube')['posse_de_bola'].mean().round(0)

#Preenchendo os valores nulos com a média da posse de bola do time correspondente
for clube, media in media_posse_de_bola.items():
    df_estatisticas.loc[(df_estatisticas['clube'] == clube) & (df_estatisticas['posse_de_bola'].isnull()), 'posse_de_bola'] = media
```

```python
#Como são muitoas linhas, e não tem como recuperar essa informação, vamos preencher com a média da posse de bola de cada time
media_posse_de_bola = df_estatisticas.groupby('clube')['precisao_passes'].mean().round(0)
for clube, media in media_posse_de_bola.items():
    df_estatisticas.loc[(df_estatisticas['clube'] == clube) & (df_estatisticas['precisao_passes'].isnull()), 'precisao_passes'] = media
```

```python
df_estatisticas.isnull().sum()
```

```python
#retirando restante de valores nulos

df_estatisticas = df_estatisticas.dropna()
```

```python
#Setando categoria data para date

#Alterando / por - para padronizar a data
df_campfull['data'] = df_campfull['data'].str.replace('/', '-')
df_campfull['data'] = pd.to_datetime(df_campfull['data'], format='%d-%m-%Y')
```

```python
#Setando dataset somente de 2020 para frente
df_campfull = df_campfull[df_campfull['ID'] >= 6888]

df_campfull.head()
```

```python
df_campfull.isnull().sum()
```

```python
#tirando coluna arrecadacao, pois tem muitos valores nulos e nao tem como recuperar essa informacao
df_campfull = df_campfull.drop(columns=['arrecadacao'])
```

```python
#Preenchendo valores nulos da coluna estadio com respectivos valores
df_campfull['formacao_mandante'] = df_campfull['formacao_mandante'].fillna('4-5-1')
df_campfull['formacao_visitante'] = df_campfull['formacao_visitante'].fillna('3-5-2')
```

```python
#Mandando respectivos datasets para processed
df_cartoes.to_csv(r'E:\Analise Brasileirao\Data\Processed\campcartoes.csv', index=False, encoding='UTF-8', sep=',')
df_estatisticas.to_csv(r'E:\Analise Brasileirao\Data\Processed\campestatisticas.csv', index=False, encoding='UTF-8', sep=',')
df_campfull.to_csv(r'E:\Analise Brasileirao\Data\Processed\campfull.csv', index=False, encoding='UTF-8', sep=',')
```
